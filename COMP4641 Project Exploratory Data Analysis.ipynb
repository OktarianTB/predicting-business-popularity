```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_json('./yelp/yelp_academic_dataset_checkin.json', lines=True)
checkins_freq_df = df[['business_id']].copy()
checkins_freq_df['frequency'] = df['date'].str.count(',')+1
checkins_freq_df.head()
```

```python
df = pd.read_json('dataset/yelp_academic_dataset_business.json', lines=True)
df = pd.merge(checkins_freq_df, df[['business_id', 'stars']], on='business_id')
df.head(3)
```

```python
user_df = pd.read_json('dataset/yelp_academic_dataset_user.json', lines=True,
                       dtype = {
                           "user_id": str,
                           "name": str,
                           "review_count": int,
                           "yelping_since": str,
                           "useful": int,
                           "funny": int,
                           "cool": int,
                           "elite": str},
                       chunksize=500000) # if you have less available memory, please change the size into a smaller one. It took me ~7.1GB at its peak at this chunksize.

# There are multiple chunks to be read
chunk_list = []
for chunk in user_df:
    chunk_list.append(chunk[['user_id', 'friends']].copy())
new_df = pd.concat(chunk_list, ignore_index=True, axis=0)
```

```python
new_df
```

```python
tmp_df = new_df.copy()
```

```python
friend_freq = tmp_df['friends'].apply(lambda x: 0 if x == 'None' else x.count(',')+1)
```

```python
fig = px.histogram(friend_freq, log_y=True)
fig.show()
```

```python
print('sum(friend_freq != 0) / len(friend_freq) =', sum(friend_freq != 0), '/', len(friend_freq), '=', sum(friend_freq != 0) / len(friend_freq))
```
